## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
sns.set(rc={"figure.figsize":(10, 6)}) 
sns.set_style('white')
sns.swarmplot(x='gender', y='age', data=kill, hue='manner_of_death', edgecolor='white')
plt.show()
```

```python
# with seaborn
```

## 2: Plot 12 most dangerous cities

```python
dang_city = kill.city.value_counts()[:12]
dang_city
```

```python
# with seaborn barplot
sns.set(rc={"figure.figsize":(8, 5)}) 
sns.set_style('white')
ax = sns.barplot(x=dang_city.index, y=dang_city.values)
ax.set_xticklabels(dang_city.index, rotation=45);
ax.set_title('Most dangerous cities',color = 'blue',fontsize=16);
```

```python
# with seaborn barplot

```

```python
# with seaborn countplot

ax= sns.countplot(x="city", data=kill,
             order=dang_city.index,
             )
ax.set_xticklabels(dang_city.index, rotation = 45)
ax.set_title('Most dangerous cities',color = 'blue',fontsize=15);
```

```python
# with seaborn countplot
```
